```python
import os
import shutil
from pathlib import Path
import yaml

def merge_and_cleanup_yolov8_datasets(dataset_folders, output_dir):
    os.makedirs(output_dir, exist_ok=True)

    for folder in ['train/images', 'valid/images', 'train/labels', 'valid/labels']:
        os.makedirs(os.path.join(output_dir, folder), exist_ok=True)

    unified_classes = []

    def load_classes(yaml_path):
        """Load class names from a YOLOv8 dataset YAML file."""
        with open(yaml_path, 'r') as f:
            data = yaml.safe_load(f)
        return data.get('names', [])

    def copy_and_update(src_img_folder, src_label_folder, dest_img_folder, dest_label_folder, class_map, folder_prefix):
        """Copy images and update labels with correct class mapping."""
        for img_file in os.listdir(src_img_folder):
            img_path = os.path.join(src_img_folder, img_file)
            label_file = img_file.replace('.jpg', '.txt').replace('.png', '.txt')
            label_path = os.path.join(src_label_folder, label_file)

            # Add folder prefix to file names
            new_img_name = f"{folder_prefix}_{img_file}"
            new_img_path = os.path.join(dest_img_folder, new_img_name)
            new_label_path = os.path.join(dest_label_folder, new_img_name.replace('.jpg', '.txt'))

            shutil.copy(img_path, new_img_path)

            if os.path.exists(label_path):
                with open(label_path, 'r') as f:
                    lines = f.readlines()

                # Update label file with the new class IDs
                with open(new_label_path, 'w') as f:
                    for line in lines:
                        parts = line.strip().split()
                        if parts:
                            class_id = int(parts[0])
                            if class_id in class_map:
                                new_class_id = class_map[class_id]
                                f.write(f"{new_class_id} {' '.join(parts[1:])}\n")
                            else:
                                print(f"Warning: Class ID {class_id} not found in class_map.")

    # Process each dataset folder
    for dataset_folder in dataset_folders:
        # Extract folder name as a prefix
        folder_prefix = Path(dataset_folder).stem

        # Load dataset classes
        yaml_path = os.path.join(dataset_folder, 'data.yaml')
        dataset_classes = load_classes(yaml_path)

        # Create a mapping from old class IDs to unified class IDs
        class_map = {}
        for i, cls in enumerate(dataset_classes):
            if cls not in unified_classes:
                unified_classes.append(cls)
            class_map[i] = unified_classes.index(cls)

        # Copy and update images and labels for train and valid
        for subset in ['train', 'valid']:
            src_img_folder = os.path.join(dataset_folder, f'{subset}/images')
            src_label_folder = os.path.join(dataset_folder, f'{subset}/labels')
            dest_img_folder = os.path.join(output_dir, f'{subset}/images')
            dest_label_folder = os.path.join(output_dir, f'{subset}/labels')

            if os.path.exists(src_img_folder) and os.path.exists(src_label_folder):
                copy_and_update(src_img_folder, src_label_folder, dest_img_folder, dest_label_folder, class_map, folder_prefix)

    # Create a new data.yaml file for the merged dataset
    merged_yaml = {
        'train': '../train/images',
        'val': '../valid/images',
        'nc': len(unified_classes),
        'names': unified_classes
    }

    with open(os.path.join(output_dir, 'data.yaml'), 'w') as f:
        yaml.dump(merged_yaml, f)

    # Delete original dataset folders
    for dataset_folder in dataset_folders:
        shutil.rmtree(dataset_folder, ignore_errors=True)

    print(f"Merged dataset created at {output_dir}")
    print(f"Unified classes: {unified_classes}")
    print("Original datasets have been deleted.")

# Example usage:
dataset_folders = [
    "./yolov8-trash-detections-6",
    "./glasses-4",
    "./mobile-phone-detection-1",
    "./BAG-DETECTION-AND-TRACKING-3",
    "./Tez-1",
    "./wallet-1"
]
output_dir = './final_dataset'

merge_and_cleanup_yolov8_datasets(dataset_folders, output_dir)
```
